```python
import numpy as np
import channel as ch
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
din = ch.read_dnsin('dns.in')
m = ch.mesh(din)
nx = din['nx']; ny = din['ny']; nz = din['nz']
nztot = 2*nz + 1
nytot = ny + 1
```

```python
cf = np.fromfile('cond_field_ejection.300_300.bin',dtype=np.float64).reshape((nztot,nytot,3))
```

```python
dz = 2*np.pi/din['beta0'] / nztot
z = np.linspace(-nz,nz,num=nztot) * dz
y = m.y[1:-1]
```

```python
fig, ax = plt.subplots()
h = ax.pcolormesh(z,y,cf[:,:,0].transpose(), shading='gouraud')
plt.colorbar(h)
stridez = slice(None, None, 40)
stridey = slice(None, None, 10)

ax.quiver(z[stridez],y[stridey],cf[stridez,stridey,2].transpose(),cf[stridez,stridey,1].transpose())
ax.set_ylim([0,0.6])
```
